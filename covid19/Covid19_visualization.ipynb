```python
# 국가코드 사전데이터인 json데이터 미리 로드 해놓기
import json

with open(path+'country_convert.json', 'r') as json_file:
    json_data = json.load(json_file)

# 국가 이름 json파일에 있는 이름에 맞게 바꾸기
def convert_country_name(row):
    if row['Country_Region'] in json_data:
        row['Country_Region'] = json_data[row['Country_Region']]
    return row

# 원하는 데이터프레임 형태로 만드는 함수 정의
def create_df(filename):
    data = pd.read_csv(dir_path + filename, encoding='utf-8')
    try:
        data = data[['Country_Region', 'Confirmed']]
    except:
        data = data[['Country/Region', 'Confirmed']]
        data.columns = ['Country_Region', 'Confirmed']
    data = data.apply(convert_country_name, axis=1)
    data = data.dropna(subset=['Confirmed'])
    data = data.astype({'Confirmed':'int64'})
    data = data.groupby('Country_Region').sum()
    
    # Confirmed 칼럼명을 파일명의 날짜로 대체하기
    date_column =filename.split('.')[0].lstrip('0').replace('-','/')
    data.columns = [date_column]
    return data
```

## 여러 데이터 수집, 전처리해서, 하나의 데이터프레임 만들기

- 지금까지의 과정을 모두 한데 모아서, 함수로 만들기
- 필요한 파일 리스트만 추출하기
- 파일 리스트 정렬하기
- 데이터프레임 전처리하기 (별도 create_dateframe() 함수)
- 데이터프레임 합치기

```python
import os 
def generate_dataframe(dir_path):
    file_list, csv_list = os.listdir(dir_path), list()
    first_doc = True
    for file in file_list:
        if file.split('.')[-1] == 'csv':
            csv_list.append(file)
        csv_list.sort()
    
    for filename in csv_list:
        doc = create_df(filename)
        if first_doc:
            final_doc, first_doc = doc, False
        else:
            final_doc = pd.merge(final_doc, doc, how='outer', left_index=True, right_index=True)
    final_doc = final_doc.fillna(0)
    final_doc = final_doc.astype('int64')
    return final_doc
```

```python
final_data = generate_dataframe(dir_path)
```

```python
merge_df.loc[merge_df['Country_Region'] == 'Comoros', 'iso2'] = 'KM'
```

```python
merge_df.loc[merge_df['Country_Region'] == 'Lesotho', 'iso2'] = 'LS'
merge_df.loc[merge_df['Country_Region'] == 'Sao Tome and Principe', 'iso2'] = 'ST'
merge_df.loc[merge_df['Country_Region'] == 'Tajikistan', 'iso2'] = 'TJ'
merge_df.loc[merge_df['Country_Region'] == 'Yemen', 'iso2'] = 'YE'
```

```python
# 크루즈 관련 결측치는 없애기
merge_df = merge_df.dropna(subset=['iso2'])
merge_df.isnull().sum()
```

```python
# 데이터프레임 칼럼명 순서 바꿔주기
cols = merge_df.columns.tolist()
cols.remove('iso2')
cols.insert(1, 'iso2')
merge_df = merge_df[cols]
merge_df.head()
```

* https://www.countryflags.io/be/shiny/64.png

```python
# 국가 이미지 링크로 변수 재생성하기 위한 함수 정의
def create_png(iso2):
    png = 'https://www.countryflags.io/'+iso2+'/shiny/64.png'
    return png
```

```python
# iso2 값에다가 함수 적용
merge_df['iso2'] = merge_df['iso2'].apply(create_png)
```

```python
# iso2 -> Flag_link로 변수명 변경
merge_df = merge_df.rename(columns={'iso2':'Flag_link'})
merge_df.head()
```
