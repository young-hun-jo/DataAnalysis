```python
import json
with open('country_convert.json', 'r') as json_file:
    json_data = json.load(json_file)

# Series 요소중에 json의 key 중에 해당하는지 조건문 할 때는 in!!
def convert_country_name(row):
    if row['Country_Region'] in json_data:
        row['Country_Region'] = json_data[row['Country_Region']]
    return row

def create_df(filename):
    data = pd.read_csv(PATH+filename, encoding='utf-8')
    try:
        data = data[['Country_Region', 'Recovered']]
    except:
        data = data[['Country/Region', 'Recovered']]
        data.columns = ['Country_Region', 'Recovered']
    data = data.dropna(subset=['Recovered'])
    data = data.astype({'Recovered':'int64'})
    data = data.apply(convert_country_name, axis=1)
    data = data.groupby('Country_Region').sum()
    
    column = filename.split('.')[0].lstrip('0').replace('-','/')
    data.columns = [column]
    return data
```

```python
def generate_df(PATH):
    csv_lists = os.listdir(PATH)
    file_lists = []
    first_doc = True
    for csv in csv_lists:
        if csv.split('.')[-1] == 'csv':
            file_lists.append(csv)
        file_lists.sort()
    
    for file in file_lists:
        doc = create_df(file)
        if first_doc:
            final_doc, first_doc = doc, False
        else:
            final_doc = pd.merge(final_doc, doc, how='outer', left_index=True, right_index=True)
    final_doc = final_doc.fillna(0)
    final_doc = final_doc.astype('int64')
    return final_doc
```

## 결측치 채워넣기

- 크루즈 관련 결측치 : Cruise Ship, Diamond Princess, MS Zaandam, Others
- iso2(국가 코드)결측치 데이터 국가 이름에 해당하는 iso2값과 mapping 시키기 from https://www.countryflags.io/
    * Coromos : KM
    * Lesotho : LS
    * Sao Tome and Principe : ST
    * Tajikistan : TJ
    * Yemen : YE

```python
final_data.loc[final_data['Country_Region'] == 'Coromos', 'iso2'] = 'KM'
final_data.loc[final_data['Country_Region'] == 'Lesotho', 'iso2'] = 'LS'
final_data.loc[final_data['Country_Region'] == 'Sao Tome and Principe', 'iso2'] = 'ST'
final_data.loc[final_data['Country_Region'] == 'Tajikistan', 'iso2'] = 'TJ'
final_data.loc[final_data['Country_Region'] == 'Yemen', 'iso2'] = 'YE'
# 크루즈선 관련 행 삭제
final_data = final_data.dropna(subset=['iso2'])
final_data.isnull().sum()
```

```python
flag_list = final_data['iso2'].tolist()
flag_link = []
for iso2 in flag_list:
    flag = 'https://www.countryflags.io/'+iso2+'/flat/64.png'
    flag_link.append(flag)
final_data['iso2'] =flag_link
final_data.head()
```

```python
# 칼럼순서 재배열
column_names = final_data.columns.tolist()
column_names.remove('iso2')
column_names.insert(1,'iso2')
final_data = final_data[column_names]
final_data = final_data.rename(columns={'iso2':'Image_URL'})
final_data.head()
```
